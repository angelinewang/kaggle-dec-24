```python
import pandas as pd
import scipy.stats as stats

# Load the dataset
data = pd.read_csv("data/train.csv")  # Replace with your dataset's filename

# Define a function to run ANOVA
def run_anova(data, categorical_col, continuous_col):
    # Group data by the categorical column
    groups = [group[continuous_col].dropna() for _, group in data.groupby(categorical_col)]
    
    # Perform ANOVA
    f_stat, p_value = stats.f_oneway(*groups)
    return f_stat, p_value

# Define columns for analysis
categorical_columns = ['Gender', 'Marital Status', 'Education Level']  # Add more if needed
continuous_columns = ['Premium Amount', 'Health Score']  # Add more if needed

# Run ANOVA for each pair
anova_results = []
for cat_col in categorical_columns:
    for cont_col in continuous_columns:
        f_stat, p_value = run_anova(data, cat_col, cont_col)
        significance = "Significant" if p_value < 0.05 else "Not Significant"
        anova_results.append({
            'Categorical Variable': cat_col,
            'Continuous Variable': cont_col,
            'F-Statistic': f_stat,
            'P-Value': p_value,
            'Significance': significance
        })

# Convert results to a DataFrame
anova_df = pd.DataFrame(anova_results)

# Display or save the results

# Option 1: Print results
print(anova_df)

# # Option 2: Save to CSV
# anova_df.to_csv("anova_results_with_significance.csv", index=False)
# print("ANOVA results saved to 'anova_results_with_significance.csv'")

# Option 3: Visualize using PrettyTable
# from prettytable import PrettyTable

# # Create PrettyTable
# table = PrettyTable()
# table.field_names = ['Categorical Variable', 'Continuous Variable', 'F-Statistic', 'P-Value', 'Significance']
# for _, row in anova_df.iterrows():
#     table.add_row(row)

# print(table)
```
